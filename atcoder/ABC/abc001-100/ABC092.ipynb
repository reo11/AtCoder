````python
%%test_input https://atcoder.jp/contests/abc092/tasks/abc092_a
# A
a = [int(input()) for i in range(4)]
ans = 0
if a[0] > a[1]:
    ans += a[1]
else:
    ans += a[0]
if a[2] > a[3]:
    ans += a[3]
else:
    ans += a[2]
print(ans)
````

````python
%%test_input https://atcoder.jp/contests/abc092/tasks/abc092_b
# B
n = int(input())
d, x = map(int, input().split())
a = [int(input()) for i in range(n)]
ans = n
for i in range(n):
    ans += (d-1) // a[i]
ans += x
print(ans)
````

````python
%%test_input https://atcoder.jp/contests/abc092/tasks/arc093_a
# C
import copy
import numpy as np
n = int(input())
a = list(map(int, input().split()))
S = sum(abs(np.array([0] + a) - np.array(a+[0])))
a = [0] + a + [0]
for i in range(1, len(a)-1):
    cost = S + abs(a[i-1] - a[i+1]) - (abs(a[i-1]-a[i]) + abs(a[i]-a[i+1]))
    print(cost)
````

````python
%%test_input https://atcoder.jp/contests/abc092/tasks/arc093_b
# D
a, b = map(int, input().split())
a -= 1
b -= 1
print("{} {}".format(100, 100))
for i in range(100):
    s = ''
    for j in range(100):
        if j < 50:
            if j % 2 == 0 and i % 2 == 0 and j < 49 and a > 0:
                s += '.'
                a -= 1
            else:
                s += '#'
        else:
            if j % 2 == 0 and i % 2 == 0 and j > 51 and b > 0:
                s += '#'
                b -= 1
            else:
                s += '.'
    print(s)
````
