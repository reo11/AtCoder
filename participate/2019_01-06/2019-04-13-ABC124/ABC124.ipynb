```python
# A
a, b = map(int, input().split())
count = 0
if a > b:
    count = 2*a - 1
elif b > a:
    count = 2*b - 1
else:
    count = 2 * a
print(count)
```

```python
# B
n = int(input())
h = list(map(int, input().split()))

count = 1
m = h[0]
for i in range(n-1):
    if m <= h[i+1]:
        count += 1
        m = h[i+1]
print(count)
```

```python
# C
s = str(input())

s = list(s)
s = list(map(lambda x: int(x), s))

count_odd = [0 for i in range(2)]
count_even = [0 for i in range(2)]

for i in range(len(s)):
    if i % 2 == 0:
        if s[i] == 0:
            count_even[0] += 1
        else:
            count_even[1] += 1
    else:
        if s[i] == 0:
            count_odd[0] += 1
        else:
            count_odd[1] += 1

ans = 0
if max(count_even) > max(count_odd):
    if count_even[0] >= count_even[1]:
        ans = count_even[1] + count_odd[0]
    else:
        ans = count_even[0] + count_odd[1]
else:
    if count_odd[0] >= count_odd[1]:
        ans = count_odd[1] + count_even[0]
    else:
        ans = count_odd[0] + count_even[1]
    
print(ans)
```

```python
# D
n, k = map(int, input().split())
s = str(input())
left = 0
right = 0
ans = 0
# 連続する0をk個分埋めたときの長さの最大値を求める
while right < n:
    count = k
    right = left
    while count != 0:
        if right >= n-1:
            break
        print(left, right)
        if s[right] != 0:
            right += 1
            continue
        else:
            count -= 1
            while s[right] == 1:
                right += 1
            continue
    ans = max(ans, right - left)
print(ans)
        
    
```

```python
# D
n, k = map(int, input().split())
s = str(input())
r = 0
l = 0
max_len = 0
count = k
ans = 0
if all([int(i) == 1 for i in list(s)]):
    ans = len(s)
else:
    while r < n:
        if r == n-1:
            break
        while count > 0:
            if r == n-1:
                break
            if int(s[r]) == 1:
                r += 1
                continue
            else:
                count -= 1
                while int(s[r]) == 0:
                    if r == n-1:
                        break
                    r += 1
                while int(s[r]) == 1:
                    if r == n-1:
                        break
                    if r == n-2 and int(s[n-1]) == 0:
                        break
                    if int(s[r+1]) == 0:
                        break
                    r += 1
                continue
        ans = max(ans, r - l + 1)
        while int(s[l]) == 1:
            if l == n-1:
                break
            l += 1
        while int(s[l]) == 0:
            if l == n-1:
                break
            l += 1
        count += 1
print(ans)
```
